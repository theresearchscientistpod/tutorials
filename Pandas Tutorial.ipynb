```python
df1 = pd.DataFrame({'id':[1,2,3,4,5],
                   'Name':['Boltzmann','Cherenkov','Dirac', 'Einstein','Feynman'],
                   'subject_id':['sub1', 'sub2', 'sub4', 'sub6', 'sub5']})

df2 = pd.DataFrame({'id':[1,2,3,4,5],
                   'Name':['Lorenz','Maxwell', 'Newton', 'Ohm', 'Planck'],
                   'subject_id':['sub2', 'sub4', 'sub3', 'sub6', 'sub5']})
```

```python
#Merge two Dataframes on a Key
pd.merge(df1, df2, on='id')
```

```python
#Merge two Dataframes on Multiple Keys
pd.merge(df1, df2, on=['id','subject_id'])
```

```python
#Left Join Using "How" Argument
pd.merge(df1, df2, on='subject_id', how='left')
```

```python
#Right Join using "How" Argument
pd.merge(df1, df2, on='subject_id', how='right')
```

```python
#Outer Join using "How" Argument
pd.merge(df1, df2, on='subject_id', how='outer')
```

```python
#Inner Join using "How" Argument
pd.merge(df1, df2, on='subject_id', how='inner')
```

```python
import plotly.express as px
import pandas as pd
```

```python
imdb_ratings = pd.read_csv('data/IMDb ratings.csv', usecols=['weighted_average_vote'])
imdb_titles = pd.read_csv('data/IMDb movies.csv', usecols=['title', 'year', 'genre'])
netflix_overall = pd.read_csv('data/netflix_titles.csv')
netflix_overall.dropna()
ratings = pd.DataFrame({'Title':imdb_titles.title,
                    'Release Year':imdb_titles.year,
                    'Rating': imdb_ratings.weighted_average_vote,
                    'Genre':imdb_titles.genre})

ratings.drop_duplicates(subset=['Title','Release Year','Rating'], inplace=True)
ratings.shape
```

```python
ratings.dropna()
merged = ratings.merge(netflix_overall, left_on='Title', right_on='title', 
                      how='inner')
merged = merged.sort_values(by='Rating', ascending=False)
```

```python
top_rated = merged[0:10]
fig = px.sunburst(top_rated,
                 path=['title', 'country'],
                 values='Rating',
                 color='Rating')
fig.show()
```
